# Moduler & Packager


```python
import datetime
import time
import math
import random
import uuid
import os

# ----------- Datetime Operations -----------
def show_current_datetime():
    now = datetime.datetime.now()
    print("Current Date and Time:", now.strftime("%Y-%m-%d %H:%M:%S"))

def date_difference():
    date1 = input("Enter the first date (YYYY-MM-DD): ")
    date2 = input("Enter the second date (YYYY-MM-DD): ")
    try:
        d1 = datetime.datetime.strptime(date1, "%Y-%m-%d")
        d2 = datetime.datetime.strptime(date2, "%Y-%m-%d")
        diff = abs((d2 - d1).days)
        print(f"Difference: {diff} days")
    except ValueError:
        print("Invalid date format!")

def format_date():
    date_str = input("Enter a date (YYYY-MM-DD): ")
    try:
        date_obj = datetime.datetime.strptime(date_str, "%Y-%m-%d")
        custom_format = date_obj.strftime("%A, %d %B %Y")
        print("Formatted Date:", custom_format)
    except ValueError:
        print("Invalid date format!")

def stopwatch():
    input("Press Enter to start the stopwatch...")
    start = time.time()
    input("Press Enter to stop...")
    end = time.time()
    elapsed = end - start
    print(f"Elapsed time: {elapsed:.2f} seconds")

def countdown_timer():
    seconds = int(input("Enter countdown time in seconds: "))
    while seconds > 0:
        print(f"Time remaining: {seconds} seconds", end='\r')
        time.sleep(1)
        seconds -= 1
    print("\nTime's up!")

# ----------- Mathematical Operations -----------
def calculate_factorial():
    n = int(input("Enter a number: "))
    print(f"Factorial: {math.factorial(n)}")

def compound_interest():
    P = float(input("Enter principal amount: "))
    R = float(input("Enter rate of interest (in %): "))
    T = float(input("Enter time (in years): "))
    A = P * (1 + R/100)**T
    print(f"Compound Interest: {A:.2f}")

# ----------- Random Data Generation -----------
def generate_random_number():
    start = int(input("Enter start range: "))
    end = int(input("Enter end range: "))
    print("Random Number:", random.randint(start, end))

def generate_random_list():
    length = int(input("Enter length of list: "))
    start = int(input("Enter start range: "))
    end = int(input("Enter end range: "))
    rand_list = [random.randint(start, end) for _ in range(length)]
    print("Random List:", rand_list)

def create_random_password():
    length = int(input("Enter password length: "))
    chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()"
    password = "".join(random.choice(chars) for _ in range(length))
    print("Generated Password:", password)

def generate_random_otp():
    length = int(input("Enter OTP length: "))
    otp = "".join(str(random.randint(0, 9)) for _ in range(length))
    print("Generated OTP:", otp)

# ----------- UUID Generation -----------
def generate_uuid():
    print("Generated UUID:", uuid.uuid4())

# ----------- File Operations -----------
def create_file():
    filename = input("Enter file name: ")
    open(filename, 'w').close()
    print(f"File '{filename}' created successfully!")

def write_file():
    filename = input("Enter file name: ")
    data = input("Enter data to write: ")
    with open(filename, 'w') as f:
        f.write(data)
    print("Data written successfully!")

def read_file():
    filename = input("Enter file name: ")
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            print("File Content:", f.read())
    else:
        print("File does not exist!")

def append_file():
    filename = input("Enter file name: ")
    data = input("Enter data to append: ")
    with open(filename, 'a') as f:
        f.write(data)
    print("Data appended successfully!")

# ----------- Module Exploration -----------
def explore_module():
    module_name = input("Enter module name to explore: ")
    try:
        mod = __import__(module_name)
        print(f"Available Attributes in {module_name} module:")
        print(dir(mod))
    except ModuleNotFoundError:
        print("Module not found!")

# ----------- Main Menu -----------
def main():
    while True:
        print("\nWelcome to Multi-Utility Toolkit")
        print("1. Datetime and Time Operations")
        print("2. Mathematical Operations")
        print("3. Random Data Generation")
        print("4. Generate Unique Identifiers (UUID)")
        print("5. File Operations")
        print("6. Explore Module Attributes")
        print("7. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            print("\nDatetime and Time Operations:")
            print("1. Display current date and time")
            print("2. Calculate difference between two dates")
            print("3. Format date into custom format")
            print("4. Stopwatch")
            print("5. Countdown Timer")
            print("6. Back to Main Menu")
            sub_choice = input("Enter your choice: ")
            if sub_choice == '1': show_current_datetime()
            elif sub_choice == '2': date_difference()
            elif sub_choice == '3': format_date()
            elif sub_choice == '4': stopwatch()
            elif sub_choice == '5': countdown_timer()

        elif choice == '2':
            print("\nMathematical Operations:")
            print("1. Calculate Factorial")
            print("2. Solve Compound Interest")
            print("3. Back to Main Menu")
            sub_choice = input("Enter your choice: ")
            if sub_choice == '1': calculate_factorial()
            elif sub_choice == '2': compound_interest()

        elif choice == '3':
            print("\nRandom Data Generation:")
            print("1. Generate Random Number")
            print("2. Generate Random List")
            print("3. Create Random Password")
            print("4. Generate Random OTP")
            print("5. Back to Main Menu")
            sub_choice = input("Enter your choice: ")
            if sub_choice == '1': generate_random_number()
            elif sub_choice == '2': generate_random_list()
            elif sub_choice == '3': create_random_password()
            elif sub_choice == '4': generate_random_otp()

        elif choice == '4':
            generate_uuid()

        elif choice == '5':
            print("\nFile Operations:")
            print("1. Create a new file")
            print("2. Write to a file")
            print("3. Read from a file")
            print("4. Append to a file")
            print("5. Back to Main Menu")
            sub_choice = input("Enter your choice: ")
            if sub_choice == '1': create_file()
            elif sub_choice == '2': write_file()
            elif sub_choice == '3': read_file()
            elif sub_choice == '4': append_file()

        elif choice == '6':
            explore_module()

        elif choice == '7':
            print("\nThank you for using the Multi-Utility Toolkit!")
            break
        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```
